```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from wordcloud import WordCloud
from sklearn.feature_extraction import text
```

```python
from nltk.corpus import stopwords
```

```python
#openning pkl-ed file with jeans information
df = pd.read_pickle("./df_all.pkl")
```

```python
df.columns
```

**_Word Cloud on reviews below 4_**

```python
reviews_bellow = df[df.rating < 4.0].review_nltk
```

```python
stop_words_add_ = (['jeans', 'jean', 'like', 'everlane', 've', 'going', 'items', 'just', 'don',
                  'order', 'past', 'couldn', 'pants','would',
                  'also', 'blue', 'kick', 'meh', 't', 'right', 'wanted', 'disappointed',
                  'unfortunately', 'see', 'get', 'hoped', 'work', 'overall', 'way',
                  'usually', 'sure', 'wish', 'one', 'good', 'however', 'look', 'is', 'pant', 'im'])
```

```python
stop_words_ = text.ENGLISH_STOP_WORDS.union(stop_words_add_)
```

```python
join_reviews_bellow = ' '.join(reviews_bellow)
```

```python
wc = WordCloud(stopwords=stop_words_, background_color="white", colormap="Dark2",
               max_font_size=150, random_state=42).generate(join_reviews_bellow.lower())
```

```python
plt.figure(dpi=200)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

**_Word Cloud on above 4 star reviews_**

```python
reviews_above = df[df.rating > 4.0].review_nltk
```

```python
stop_words_add_positive = (['jeans', 'jean', 'pant', 'everlane', 've', 'going', 'items', 'just', 'don',
                  'order', 'past', 'couldn', 'pants','would',
                  'also', 'blue', 'kick', 'meh', 't', 'right', 'wanted', 'disappointed',
                  'unfortunately', 'see', 'get', 'hoped', 'work', 'overall', 'way', 'ha',
                  'usually', 'sure', 'wish', 'one', 'however', 'look', 'is', 'are', 'bought', 'pair',
                  'fit', 'high', 'rise', 'wear', 'waist', 'skinny', 'size', 'im', 'denim', 'theyre'])
```

```python
stop_words_positive = stopwords.words('english')

#adding new words to stop words list
stop_words_positive.extend(stop_words_add_positive)
```

```python
join_reviews_above = ' '.join(reviews_above)
```

```python
wc_positive = WordCloud(stopwords=stop_words_positive, background_color="white", colormap="Dark2",
               max_font_size=150, random_state=42).generate(join_reviews_above.lower())
```

```python
plt.figure(dpi=200)
plt.imshow(wc_positive, interpolation='bilinear')
plt.axis("off")
plt.show()
```
